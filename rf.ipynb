```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df.head()
```

```python
def sample_rows(df,percent):
    return df.sample(int(percent*df.shape[0]),replace=True)

```

```python
df1=sample_rows(df,0.1)
df2=sample_rows(df,0.1)
df3=sample_rows(df,0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1=DecisionTreeClassifier()
clf2=DecisionTreeClassifier()
clf3=DecisionTreeClassifier()
```

```python
clf1.fit(df1.iloc[:,0:5],df1.iloc[:,-1])
clf2.fit(df1.iloc[:,0:5],df2.iloc[:,-1])
clf3.fit(df1.iloc[:,0:5],df3.iloc[:,-1])
```

```python
import random

def sample_columns(df, percent):
    # Convert columns to list, exclude last column
    column_list = df.columns.tolist()[:-1]

    # Number of columns to sample
    num_cols = int(percent * len(column_list))

    # Randomly pick columns
    cols = random.sample(column_list, num_cols)

    return df[cols]
```

```python
df1=sample_columns(df,0.8)
```

```python
df2=sample_columns(df,0.8)
```

```python
df3=sample_columns(df,0.8)
```
