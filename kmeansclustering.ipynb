```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
##ELBOW METHOD to select he k value
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    score=silhouette_score(X_train_scaled,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficients")
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Age': [19,21,20,23,31,35,40,50,45,60],
    'Income': [15000,18000,16000,19000,55000,60000,65000,70000,72000,80000]
}

df = pd.DataFrame(data)

# KMeans model
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit model
df['cluster'] = kmeans.fit_predict(df[['Age', 'Income']])

print(df)
```

```python
plt.scatter(df['Age'], df['Income'], c=df['cluster'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Segmentation using K-Means")
plt.show()
```
